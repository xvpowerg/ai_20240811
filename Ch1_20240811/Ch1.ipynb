```python
#RD:0 HR:1 SA:2 PM:3
type = input("工作類型:")
result = -1
if type == "RD":
    result = 0
else:
    if type == "HR":
        result = 1
    else:
        if type == "SA":
            result = 2
        else:
            if type == "PM":
                result = 3
            else:
                print("錯誤的類型")
print(type,result)                
```

```python
aqi = int(input("輸入AQI指標"))
if aqi >= 0 and aqi <= 50:
    print("良好")
elif  51 <=  aqi <= 100:
    print("普通")
elif aqi >= 101 and aqi <= 150:
    print("對敏感族群不健康")
elif 151 <= aqi  <= 200:
    print("對所有群不健康")
elif aqi >= 201:
    print("非常不健康")
else:
    print("錯誤")
```

```python
# BMI = 體重 / 身高^2 (單位公尺)
h = float(input("請輸入身高"))
w = int(input("請輸入體重"))
bmi = w / (h / 100) ** 2
print(bmi)
print(round(bmi,2))
print(f"{bmi:.2f}")
```
