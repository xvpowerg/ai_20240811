## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。


```python
str1 = "This is Python.That is Java,This is SQLite"
idx =  str1.find("Python")
print(idx)
idx2 = str1.find("Java")
print(idx2)
idx3 = str1.find("is")
print(idx3)
idx4 = str1.find("not")
print(idx4)

idx5 =  str1.find("is",3)
print(idx5)
```

## 找出所有is的index

```python
search1 = "is"
index = 0
while True:
    index = str1.find(search1,index)
    if index == -1:
        break
    print(index)
    index += 1
```

# List VS Tuple

+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改
## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。

```python
myList = ["H","I","K"]
tuple1 = ("A","B","C","D")

print(len(myList))
print(len(tuple1))

print(myList)
print(tuple1)
myList[1] = 77
print(myList)
#tuple1[2] = 99
#print(tuple1)
```

```python
myList = ["Apple"]
print(type(myList))
print(myList)
myTuple = ("Banana")
print(type(myTuple))
print(myTuple)
myTuple = ("Banana",)
print(type(myTuple))
print(myTuple)
```

```python
myList = ["A","B"]
myList.append("C")
myList.append("D")
print(myList)
```

```python
## 輸入姓名一群姓名 直到輸入n表示完成
# Ken
# Sean
# Vivin
# n
# Ken,Sean,Vivin
names = []
while True:
    name = input("請輸入姓名完成請按n")
    if name == "n":
        break
    names.append(name)    

nLen = len(names)
for i in range(nLen):
    print(names[i],end = "")
    if i < nLen -1:
        print("",end =",")
        
newStr = ",".join(names)
print()
print(newStr)
```

## list 項目插入新資料
+ list 可使用 insert 方式插入新的資料：
   + list.insert(索引值，插入的資料)。
   + 若索引值超出範圍則會改為 append 方式附加。

```python
myList = ['a',"b",'c',"d",'e']
print(myList)
print(len(myList))
myList.insert(1,"x")
print(myList)
print(len(myList))
myList.insert(-2,"y")
print(myList)
print(len(myList))
myList.insert(100,'g')
print(myList)
print(len(myList))
```
