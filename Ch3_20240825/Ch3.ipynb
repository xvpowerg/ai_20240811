## 作業 文字接龍

```python
txt = input("請輸入第一個字串")
strList = []
strList.append(txt)
msg = txt
count = 0
inputMsg = "請輸入-"+txt[-1]+"開始的字串"
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = "請輸入-"+txt[-1]+"開始的字串"
    else:
        inputMsg = "請再次輸入-"+txt[-1]+"開始的字串"
        count+=1
    print("上個字串是:",msg)    
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[2]
print(list2)
del list2[1:3]
print(list2)
del list2[3:5]
print(list2)
list3 = []
del list3[1:5]
```

## isinstance

isinstance(x,int) 代表判斷x是否為int型態

list若要進行排序得先挑選出相同型態資料。

請使用isinstance(資料,資料型態)方式於list中撈出特定資料。

list2a=[x for x in list2 if isinstance(x,int)]

這樣的寫法的目的

避免程式流程受到外面的干擾

流程單純化，list的資料變化就在list區域內進行

促進語法的可讀性提高!


```python
x = 2
if isinstance(x,int):
    print("x is int")
else:
    print("x is not int")    
```

```python
myList = ["x",1,"u",2,5,"z"]
myIntList = [v for v in myList if  isinstance(v,int) ]
print(myIntList)
#顯示字串List
myStrList = [v for v in myList if isinstance(v,str)]
print(myStrList)
```

## Dictionary

資料結構是由 key:value 所組成。

key 不能夠重複，否則會被後面的結果蓋過去。

可輸入 key 找尋您要找出來的值。

如果輸入的 key 不存在，那就會出現錯誤訊息。

key 只能是 字串 數字 tuple

```python
dict1 =  {"a":100,"b":200,"c":300}
print(dict1)
print(dict1["b"])
print(dict1["x"])
```

```python
dict1 =  {"a":100,"b":200,"c":300}
for key in dict1.keys():
    print(key)
print("==========================")    
for v in dict1.values():
    print(v)
print("==========================")    
for k in dict1:
    print(k)
```

## 自訂函數的運用

減少重複語法對於之後的維修會比較方便。


#### 要如何減少重複的相同語法?

把重複的相同語法當作工具看待，需要的時候呼叫他。

這種工具於程式語法內叫做函數(function)。

```python
for i in range(1,11):
    print(i,end = " ")
print()
for i in range(5,10):
    print(i,end=" ")
```

```python
def func_sum():
    print("呼叫函數")

func_sum()
```

### 自訂函數接收資料：

傳入函數的參數放在 ( ) 內。

可以接收多個，以逗點隔開。

接收參數可以是變數，也可以是 list。

函數若設定接收參數，呼叫函數時一定要給參數。

接收參數可設定為不固定數量。

若不接收則為空白。


```python
def func_sum2(a,b):
    c = a + b
    print(c)
func_sum2(2,3)    
```

```python
def forLoop(a,b):
    if a < b:        
        for i in range(a,b+1):
            print(i,end=" ")
    else:
        for i in range(a,b-1,-1):
            print(i,end=" ")
    print()
    
forLoop(1,10)    
forLoop(5,9)
forLoop(9,5)#9 8 7 6 5 
```

```python
def forLoop(a,b):
    offset = 1  if a < b  else -1
    for i in range(a,b + offset,offset):
        print(i,end = " ")
    print()   
    
forLoop(1,10)    
forLoop(5,9)
forLoop(9,5)   
```
