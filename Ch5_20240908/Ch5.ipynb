## 輸入四個月的支出金額後列出最多與最少的支出金額

```python
projList = []
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額"))
    priceSum += p
    projList.append(p)
projList.sort()

print(f"Max:{projList[-1]}")
print(f"Min:{projList[0]}")    
print(projList)
print(f"總和:{priceSum}")
```

```python
projList = []
priceSum = 0
count = 0
while count < 4:
    
    count+=1
    try:
        inputStr = input(f"請輸入{count}個月的支出金額")
        p = int(inputStr)
    except:
        print(f"輸入錯誤的金額{inputStr}")
        count-=1
        continue
    
    priceSum += p
    projList.append(p)
 
projList.sort()

print(f"Max:{projList[-1]}")
print(f"Min:{projList[0]}")    
print(projList)
print(f"總和:{priceSum}")
```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
str1 = "Ken,Vivin,Lucy"
nameList = str1.split(",")
print(nameList[0])
print(nameList[1])
print(nameList[2])
```

```python
import re
text = 'one, two...ten'
re1 = re.split("[,. ]+",text)
print(re1)
```

```python
str1 = "Ken:80,Vivin:70,Lucy:95"
#希望幫我輸出Ken 分數:80 
# Vivin 分數:70
# Lucy 分數:95
dataList = str1.split(",")
for data in dataList:
    finalData = data.split(":")
    print(finalData[0] ,"分數:",finalData[1])
```

## sub 方法
+ sub 方法用於替換，它的使用形式如下
  + sub(取代的字串或函數,來源字串,可設定被取代的數量)
+ 可設定被取代的數量若不指定時全部替換。

```python
import re
inputStr = "hello 123 world 456"
rStr = re.sub(r"\d+","222",inputStr)
print(rStr)
```

```python
#circle 
#建立時必須提供半徑r
#計算面積 area pi * r **２
#與周長 length 2 * pi * r
class Circle:
    pi = 3.1415
    def __init__(self,r):
        self.r = r
    def area(self):
        return Circle.pi * self.r ** 2 
    def length(self):
        return Circle.pi * self.r * 2
c = Circle(10)
print(round(c.area(),2) )
print(f"{c.length():.2f}" )
```

## 繼承

```python
class Animal:
    def __init__(self,name,age):
        self.__name = name
        self.age = age
    def setName(self,name):
        self.__name = name
    def getName(self):
        return self.__name
    def printInfo(self):
        print(f"{self.getName()}:{self.age}")
a1 = Animal("Bobo",2)
a1.printInfo()
a1.setName("Kiki")
a1.printInfo()
```

```python
class Animal:
    def __init__(self,name,age = 10):
        self.name = name
        self.age = age
    def run(self):
        print(self.name,"跑")
    def eat(self):
        print(self.name,"吃")
class Cat(Animal):
    def accelerate(self):
        print("貓加速:",end="")
        self.run()
class  Bird(Animal):
    def fly(self):
        print("鳥會飛")
     ##複寫 Override   
    def run(self):
        print("鳥會跑")
        #呼叫父類別的run
        super().run()
cat1 = Cat("Gigi",3)
cat1.accelerate()
cat1.run()
bird1 = Bird("Bobo",2)
bird1.fly()
bird1.run()
```
