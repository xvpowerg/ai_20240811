## 變數影響範圍
+ 函數外的變數：
  + 函數內可以顯示該變數內容
  + 不屬於函數的區域內都可以使用
+ 函數內的變數：
  + 只在函數內產生效果，不會影響函數外的變數
  + 若函數內沒有進行變數宣告而進行改變內容動作將會產生錯誤訊息!

```python
def calculate(*values,**args):
    cType = "+"
    if "cType" in args:
        cType = args["cType"]
    result = 0    
    if cType == "*":
       result = 1
    for v in values:
        if cType == "+":
            result += v
        elif cType == "*":
            result *= v
    return result        
result = calculate(1,2,3,4,cType = "+")
result2 = calculate(1,2,3,4,cType = "*")
print(result)
print(result2)
```

```python
def myPrint(*values,**args):    
    sep = " "
    end = "\n"
    dataLen = len(values)
    if "end" in args:
        end = args["end"]
    if "sep" in args:
        sep = args["sep"]
        
    for i in range(dataLen):
        print(values[i],sep="",end = "")
        if i < dataLen - 1:
         print(sep,sep="",end="")
    print(end,sep="",end="")

myPrint("A","B","C","D",sep=",")
myPrint("X",end="@")
myPrint("Y",end="@")
myPrint("Z")
myPrint("KL")
```

## 匿名函數
不需要定義函數名稱，只需要用到運算式或表達分析語法。

Python使用lambda語法定義匿名函數。

匿名函數是一個表達式或計算式，並不是一個執行流程區塊。

匿名函數出現在一般函數不允許的地方，例如list裡面或著函數呼叫參數的位置。

```python
def f(x,y,z):
    return x+y+z
print(f(2,30,400)) 
f2 = lambda x,y,z : x + y + z
print(f(2,30,400))
print(f2(1,2,3))
print(type(f))
print(f2)
```

```python
def f1(x):
    return x ** 2
def f2(x):
    return x ** 3
def f3(x):
    return x ** 4
funList = [f1,f2,f3]    
for f in funList:
    print(f(3))
```

```python
list2 = [lambda x: x**2,lambda x:x**3,lambda x:x**4]
for f in list2:
    print(f(3))
```

## 例外語法
try:
    保護的程式區段
    
except 例外1 :
    例外1發生時執行的程式
    
except :
    其他例外發生時執行的程式
    #未指定例外型態，捕捉所有例外物件
    
else:
    若try程式區段沒產生例外執行的程式
    
finally:
    不管有沒有發生例外都會執行的程式

```python
num1 = 10
num2 = 0

print("Start")
try:
    print(num1 / num2)
except:
    print("錯誤發生")

print("End")
```

```python
num1 = 10
num2 = 1
nums = [1,2,3,4]

num2 = int(input("請輸入整數"))
print("Start")
try:
    print(num1 / num2)
    print(nums[num2])
except ZeroDivisionError :
    print("分母不可為0")
except IndexError:
    print("Index超過List範圍")
print("End")
```

```python
num1 = 10
num2 = 1

try:
    print(num1 / num2)
except:
    print("發生錯誤")
else:
    print("無例外發生")
finally:
    print("一定會執行")
```

## 使用raise拋例外

```python
## 某個函數可檢測成績testScore
## 函數傳入score 範圍必須在0~100之間
## 不在0~100顯示錯誤的成績
def testScore(score):
    if score < 0 or score >100:
        print("錯誤的成績")
        return False
    return True
       
print(testScore(100))      
print(testScore(-1))   
print(testScore(25))  
```

```python
def testScore(score):
    if score < 0 or score >100:
        raise OverflowError("錯誤的成績")
    return True
       
print(testScore(100))      
print(testScore(-1))   
print(testScore(25))
```

```python
def testScore(score):
    if score < 0 or score >100:
        raise OverflowError(f"錯誤的成績:{score}")
    return True
try:
    print(testScore(100))      
    print(testScore(-1))   
    print(testScore(25))
except OverflowError as ex:
    print("錯誤的成績:",ex)

```

```python
def changeList(myList):
    myList[0] = 100
myList = [15]
print(myList[0])
changeList(myList)
print(myList[0])
```
